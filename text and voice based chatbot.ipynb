```python

import speech_recognition as sr
import pyttsx3
import googlesearch

# Initialize speech recognition and synthesis engines
recognizer = sr.Recognizer()
synthesizer = pyttsx3.init()

# Define a simple chatbot function
def chatbot(text):
    if "hello" in text:
        return "Hello! How can I assist you?"
    elif "how are you" in text:
        return "I'm just a computer program, so I don't have feelings, but thank you for asking!"
    elif "what is your name" in text:
        return "my name is Gangaraju, i'm a chatbot,how can i help you?."
    elif "What government schemes and subsidies are available for farmers in India" in text:
        return " The Government of India offers various schemes and subsidies to support farmers, including schemes for crop insurance, soil health cards, agricultural loans, and subsidies on seeds and fertilizers!." 
    elif "What kinds of fertilizers do you offer" in text:
        return " We offer a wide range of fertilizers including organic, synthetic, and specialty blends tailored to various crops and soil types. Feel free to ask for recommendations based on your specific needs!."
    elif "Can I get information on crop cultivation and best agricultural practices?" in text:
        return "Yes, the government provides advisory services to farmers regarding crop selection, cultivation techniques, pest management, and post-harvest practices. You can access this information through agricultural extension services and mobile apps!."
    elif  "How can I get weather forecasts and alerts for my region?" in text:
        return "The Government of India provides weather forecasting services through various agencies. You can receive weather updates and alerts through mobile apps, websites, and agricultural extension offices to plan your farming activities accordingly!."
    elif "Where can I find information on market prices for agricultural produce?" in text:
        return  "The government offers market information services to help farmers track prices of agricultural commodities in local markets and wholesale markets across India. You can access this information through mobile apps, websites, and dedicated helplines!."
    elif "Are there any training programs available for farmers to improve their skills?" in text:
        return "Yes, the Government of India organizes various training and skill development programs for farmers to enhance their knowledge and skills in agriculture, livestock management, horticulture, and allied sectors. You can inquire about upcoming programs and workshops through agricultural extension offices and online portals!."
    elif "How can I avail financial assistance and subsidies for agricultural equipment?" in text:
        return "The government offers financial assistance and subsidies to farmers for the purchase of agricultural machinery and equipment such as tractors, pumps, and drip irrigation systems. You can apply for subsidies through government schemes and agricultural departments!."
    elif "How do you get this Idea ,to Help farmers?" in text:
        return "This idea or thougt i got from the under the guidance of Manjunath sir,he is a faculty in presidency university in the branch of computer scince,he is approch to do the chatbot for kissan application to help  farmers! Thank you sir! ."
    elif "bye" in text:
        return "Goodbye! Have a great day!"
    else:
            # Perform Google search for the query
        search_results = googlesearch.search(text, num=3, stop=3, pause=2)
        response = "I found the following results:\n"
        for i, result in enumerate(search_results, 1):
            response += f"{i}. {result}\n"
        return response



# Function to recognize speech
def recognize_speech():
    with sr.Microphone() as source:
        print("Listening...")
        recognizer.adjust_for_ambient_noise(source)  # Adjust for ambient noise
        audio = recognizer.listen(source)  # Listen for speech
        print("Recognizing...")
        try:
            text = recognizer.recognize_google(audio)  # Recognize speech using Google Speech Recognition
            print("You said:", text)
            return text.lower()
        except sr.UnknownValueError:
            print("Sorry, I couldn't understand the audio.")
            return None
        except sr.RequestError as e:
            print("Error fetching results; {0}".format(e))
            return None

# Function to get text input from the user
def get_text_input():
    return input("Enter your message: ")

# Function to synthesize speech
def speak(text):
    synthesizer.say(text)
    synthesizer.runAndWait()

# Main loop for chatbot
if __name__ == "__main__":
    while True:
        print("Choose input option:")
        print("1. Speech")
        print("2. Text")
        choice = input("Enter your choice (1 or 2): ")
        
        if choice == "1":  # Speech input
            text = recognize_speech()
            if text:
                response = chatbot(text)
                print("Bot:", response)
                speak(response)
        elif choice == "2":  # Text input
            text = get_text_input()
            response = chatbot(text)
            print("Bot:", response)
            speak(response)
        else:
            print("Invalid choice. Please enter 1 or 2.")

```
